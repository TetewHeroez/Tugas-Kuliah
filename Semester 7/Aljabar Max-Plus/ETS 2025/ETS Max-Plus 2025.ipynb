### No. 1  
Diberikan matriks  $A = \begin{bmatrix}-1 & 11 \\1 & \varepsilon\end{bmatrix}$ Tunjukkan bahwa untuk $k \ge 2$, maka $A^{\otimes (k+2)} = \lambda^2 \otimes A^{\otimes k},$ di mana $\lambda$ adalah nilai eigen dari matriks $A$.

### Solusi:

Perhatikan pola perkalian matriks max-plus berikut:
\begin{align*}
A^{\otimes 2} &= \begin{bmatrix}-1 & 11 \\1 & \varepsilon\end{bmatrix} \otimes \begin{bmatrix}-1 & 11\\1 & \varepsilon\end{bmatrix} = \begin{bmatrix}\max(-1-1,11+1) & \max(-1+11,11+\varepsilon)\\\max(1-1,\varepsilon+1) & \max(1+11,\varepsilon+\varepsilon)\end{bmatrix}=\begin{bmatrix}12 & 10\\0 & 12\end{bmatrix}. \\
A^{\otimes 3} &= A^{\otimes 2} \otimes A =  \begin{bmatrix}12 & 10 \\0 & 12\end{bmatrix} \otimes \begin{bmatrix}-1 & 11 \\1 & \varepsilon\end{bmatrix} = \begin{bmatrix}\max(12-1,10+1) & \max(12+11,10+\varepsilon)\\\max(0-1,12+1) & \max(0+11,12+\varepsilon)\end{bmatrix}=\begin{bmatrix}11 & 23 \\13 & 11\end{bmatrix}. \\
A^{\otimes 4} &= A^{\otimes 3} \otimes A = \begin{bmatrix}11 & 23 \\13 & 11\end{bmatrix} \otimes \begin{bmatrix}-1 & 11 \\1 & \varepsilon\end{bmatrix} = \begin{bmatrix}\max(11-1,23+1) & \max(11+11,23+\varepsilon)\\\max(13-1,11+1) & \max(13+11,11+\varepsilon)\end{bmatrix}=\begin{bmatrix}24 & 22 \\12 & 24\end{bmatrix}. \\
A^{\otimes 5} &= A^{\otimes 4} \otimes A = \begin{bmatrix}24 & 22 \\12 & 24\end{bmatrix} \otimes \begin{bmatrix}-1 & 11 \\1 & \varepsilon\end{bmatrix} = \begin{bmatrix}\max(24-1,22+1) & \max(24+11,22+\varepsilon)\\\max(12-1,24+1) & \max(12+11,24+\varepsilon)\end{bmatrix}=\begin{bmatrix}23 & 35 \\25 & 23\end{bmatrix}. \\
\end{align*}

Dari perhitungan di atas, dapat dilihat bahwa pola dari $A^{\otimes k}$ untuk $k \ge 2$ adalah sebagai berikut:
\begin{align*}
A^{\otimes 4} &= \begin{bmatrix}24 & 22 \\12 & 24\end{bmatrix} = 12 \otimes \begin{bmatrix}12 & 10 \\0 & 12\end{bmatrix} = \lambda^2 \otimes A^{\otimes 2}, \\
A^{\otimes 5} &= \begin{bmatrix}23 & 35 \\25 & 23\end{bmatrix} = 12 \otimes \begin{bmatrix}11 & 23 \\13 & 11\end{bmatrix} = \lambda^2 \otimes A^{\otimes 3},
\end{align*}
Artinya $\lambda^{\otimes 2} = 12 \implies \lambda = 6$. Selanjutnya akan dibuktikan menggunakan induksi matematika bahwa ekspresi $A^{\otimes (k+2)} = \lambda^{\otimes 2} \otimes A^{\otimes k}$ berlaku untuk semua $k \ge 2$.

- **Basis Induksi:** Untuk $k=2$, telah dibuktikan di atas bahwa $A^{\otimes 4} = \lambda^{\otimes 2} \otimes A^{\otimes 2}$.
- **Asumsi Induksi:** Misalkan untuk suatu $k \ge 2$, berlaku $A^{\otimes (k+2)} = \lambda^{\otimes 2} \otimes A^{\otimes k}$.
- **Langkah Induksi:** Perlihatkan bahwa $A^{\otimes (k+3)} = \lambda^{\otimes 2} \otimes A^{\otimes (k+1)}$.
\begin{align*}
A^{\otimes (k+3)} &= A^{\otimes (k+2)} \otimes A \\
&= (\lambda^{\otimes 2} \otimes A^{\otimes k}) \otimes A \quad \text{(dari asumsi induksi)} \\
&= \lambda^{\otimes 2} \otimes (A^{\otimes k} \otimes A) \\
&= \lambda^{\otimes 2} \otimes A^{\otimes (k+1)}.
\end{align*}

Dengan demikian, berdasarkan prinsip induksi matematika, telah terbukti bahwa untuk semua $k \ge 2$, berlaku $A^{\otimes (k+2)} = \lambda^{\otimes 2} \otimes A^{\otimes k}$.


### No. 2  
Diberikan $a, b, c \in \mathbb{R}$ dan matriks  

$$
A = 
\begin{bmatrix}
a & b \\
\varepsilon & c
\end{bmatrix}.
$$

Berikan apa **syarat nilai** $a, b, c$ dari $A$ supaya matriks $A$ mempunyai nilai eigen yang sama dengan $c$. Buktikan jawaban saudara.

### Solusi:

$\lambda = c$ adalah nilai eigen dari matriks $A$ jika dan hanya jika terdapat vektor eigen $x \neq \varepsilon$ sehingga 
$$
A \otimes x = \lambda \otimes x.
$$
Misalkan $x = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$ dengan $x_1, x_2 \in \mathbb{R}$. Maka,
\begin{align*}
\begin{bmatrix}
a & b \\
\varepsilon & c
\end{bmatrix} \otimes \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} &= \begin{bmatrix}
\max(a + x_1, b + x_2) \\
\max(\varepsilon + x_1, c + x_2)
\end{bmatrix} = \begin{bmatrix}
c + x_1 \\
c + x_2
\end{bmatrix}.
\end{align*}
Dari persamaan di atas, diperoleh dua persamaan:
\begin{align*}
\max(a + x_1, b + x_2) &= c + x_1, \\
\max(\varepsilon + x_1, c + x_2) &= c + x_2.
\end{align*}
Persamaan kedua selalu terpenuhi karena $\varepsilon + x_1 = \varepsilon$. Oleh karena itu, kita hanya perlu mempertimbangkan persamaan pertama:
\begin{align*}
\max(a + x_1, b + x_2) &= c + x_1.
\end{align*}
Menurut sifat maksimum, persamaan ini dapat dipecah menjadi dua kasus:
\begin{align*}
\begin{cases}
a + x_1 &\leq c + x_1, \\
b + x_2 &\leq c + x_1.
\end{cases}
\end{align*}
Dari kasus pertama, kita mendapatkan syarat $a \leq c$. Selanjutnya untuk kasus kedua didapatkan syarat $b + x_2 \leq c + x_1$ yang dapat ditulis ulang menjadi $b - c \leq x_1 - x_2$.

Karena $x_1$ dan $x_2$ dapat dipilih secara bebas (dengan syarat tidak keduanya $\varepsilon$), maka hal ini secara tidak langsung menyiratkan bahwa $b$ bisa bernilai berapapun selama $a \leq c$ terpenuhi.

Jadi, syarat nilai $a, b, c$ dari matriks $A$ supaya mempunyai nilai eigen yang sama dengan $c$ adalah:
- $a \leq c$,
- $b \in \mathbb{R}$ (bebas).

### No. 3  
Dengan menggunakan aturan **Cramer**, selesaikan sistem persamaan $A\otimes x = b$ dan tentukan mana yang mempunyai penyelesaian tunggal serta mana yang mempunyai penyelesaian banyak, beserta alasannya. untuk

- $ A = \begin{bmatrix} 1 & -9 & 4 \\ -4 & 18 & -8 \\ 2 & 1 & -4 \end{bmatrix}, \quad b = \begin{bmatrix} 1 \\ -6 \\ -3 \end{bmatrix} $

- $ A = \begin{bmatrix} 1 & -9 & 4 \\ -4 & 18 & -8 \\ 2 & 1 & -4 \end{bmatrix}, \quad b = \begin{bmatrix} 2 \\ 1 \\ 3 \end{bmatrix} $

### Solusi:
- Untuk poin (a), diperoleh
\begin{align*}
\det(A) =\begin{vmatrix} 1 & -9 & 4 \\ -4 & 18 & -8 \\ 2 & 1 & -4 \end{vmatrix} &= 1 \otimes \begin{vmatrix} 18 & -8 \\ 1 & -4 \end{vmatrix} \oplus (-9) \otimes \begin{vmatrix} -4 & -8 \\ 2 & -4 \end{vmatrix} \oplus 4 \otimes \begin{vmatrix} -4 & 18 \\ 2 & 1 \end{vmatrix} \\
&= 1 \otimes \max(18-4, -8+1) \oplus (-9) \otimes \max(-4-4, -8+2) \oplus 4 \otimes \max(-4+1, 18+2) \\&= 1 \otimes 14 \oplus (-9) \otimes -6 \oplus 4 \otimes 20 = 15 \oplus -15 \oplus 24 = 24.
\end{align*}
Selanjutnya, hitung determinan dari matriks yang diperoleh dengan mengganti kolom ke-$i$ dari $A$ dengan vektor $b$ untuk $i = 1, 2, 3$:
\begin{align*}
\det(A_1) &= \begin{vmatrix} 1 & -9 & 4 \\ -6 & 18 & -8 \\ -3 & 1 & -4 \end{vmatrix} = 1 \otimes \max(18-4, -8+1) \oplus (-9) \otimes \max(-6-4, -8-3) \oplus 4 \otimes \max(-6+1, 18-3) = 15 \oplus -19 \oplus 19 = 19, \\
\det(A_2) &= \begin{vmatrix} 1 & 1 & 4 \\ -4 & -6 & -8 \\ 2 & -3 & -4 \end{vmatrix} = 1 \otimes \max(-6-4, -8-3) \oplus 1 \otimes \max(-4-4, -8+2) \oplus 4 \otimes \max(-4-3, -6+2) = -9 \oplus -5 \oplus -0 = 0, \\
\det(A_3) &= \begin{vmatrix} 1 & -9 & 1 \\ -4 & 18 & -6 \\ 2 & 1 & -3 \end{vmatrix} = 1 \otimes \max(18-3, -6+1) \oplus (-9) \otimes \max(-4-3, -6-3) \oplus 1 \otimes \max(-4+1, 18+2) = 16 \oplus -16 \oplus 21 = 21.
\end{align*}
Maka, solusi dari sistem persamaan $A \otimes x = b$ adalah
\begin{align*}
x_1 &= \det(A_1) - \det(A) = 19 - 24 = -5, \\
x_2 &= \det(A_2) - \det(A) = 0 - 24 = -24, \\
x_3 &= \det(A_3) - \det(A) = 21 - 24 = -3.
\end{align*}
Jadi, solusi dari sistem persamaan adalah $x = \begin{bmatrix} -5 \\ -24 \\ -3 \end{bmatrix}$. Karena determinan $\det(A) \neq \varepsilon$, maka sistem persamaan ini mempunyai penyelesaian tunggal.

- Untuk poin (b), diperoleh
\begin{align*}
\det(A) =\begin{vmatrix} 1 & -9 & 4 \\ -4 & 18 & -8 \\ 2 & 1 & -4 \end{vmatrix} &= 24 \quad \text{(dari perhitungan sebelumnya)}.
\end{align*}
Selanjutnya, hitung determinan dari matriks yang diperoleh dengan mengganti kolom ke-$i$ dari $A$ dengan vektor $b$ untuk $i = 1, 2, 3$:
\begin{align*}
\det(A_1) &= \begin{vmatrix} 2 & -9 & 4 \\ 1 & 18 & -8 \\ 3 & 1 & -4 \end{vmatrix} = 2 \otimes \max(18-4, -8+1) \oplus (-9) \otimes \max(1-4, -8+3) \oplus 4 \otimes \max(1+1, 18+3) = 16 \oplus -14 \oplus 25 = 25, \\
\det(A_2) &= \begin{vmatrix} 1 & 2 & 4 \\ -4 & 1 & -8 \\ 2 & 3 & -4 \end{vmatrix} = 1 \otimes \max(1-4, -8-3) \oplus 2 \otimes \max(-4-4, -8+2) \oplus 4 \otimes \max(-4+3, 1+2) = -2 \oplus -4 \oplus 7 = 7, \\
\det(A_3) &= \begin{vmatrix} 1 & -9 & 2 \\ -4 & 18 & 1 \\ 2 & 1 & 3 \end{vmatrix} = 1 \otimes \max(18+3, 1+1) \oplus (-9) \otimes \max(-4+3, 1-3) \oplus 2 \otimes \max(-4+1, 18+1) = 22 \oplus -10 \oplus 22 = 22^v.
\end{align*}
Karena terdapat elemen \textit{ghost} (ditandai dengan simbol $^v$) pada perhitungan $\det(A_3)$, maka solusi dari sistem persamaan $A \otimes x = b$ tidak dapat ditentukan secara unik. Hal ini menunjukkan bahwa sistem persamaan ini mempunyai penyelesaian banyak.

Jadi, solusi dari sistem persamaan adalah $x = \begin{bmatrix} 1 \\ -17 \\ x_3 \end{bmatrix}$ dengan $x_3\leq -2$.
