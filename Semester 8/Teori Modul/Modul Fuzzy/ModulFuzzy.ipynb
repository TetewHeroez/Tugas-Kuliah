## BAB II: LANDASAN TEORI

### 2.1 Konsep Dasar Modul
Misalkan $R$ adalah sebuah gelanggang (*ring*) dengan elemen identitas $1_R$. Sebuah $R$-modul (tepatnya modul kiri) $M$ adalah suatu grup abelian $(M, +)$ yang dilengkapi dengan operasi perkalian skalar $\cdot : R \times M \to M$ yang memenuhi aksioma berikut $\forall r, s \in R$ dan $\forall x, y \in M$:
1. $r \cdot (x + y) = r \cdot x + r \cdot y$
2. $(r + s) \cdot x = r \cdot x + s \cdot x$
3. $(rs) \cdot x = r \cdot (s \cdot x)$
4. $1_R \cdot x = x$

*Remark: Modul adalah perumuman abstrak dari ruang vektor di mana skalarnya diambil dari gelanggang, bukan lapangan. (Bland, 2011)*

### 2.2 Objek Lanjut dalam Teori Modul (Prasyarat Nakayama)
Untuk memahami perilaku "penyerapan anomali" pada pengamatan kita secara sistematis, beberapa objek aljabar terstruktur wajib diperkenalkan terlebih dahulu agar tidak menjadi *black-box*:
1. **Submodul & Pembangun Berhingga (*Finitely Generated*):** Subhimpunan $N \subseteq M$ disebut submodul jika ia memiliki struktur kelompok yang tertutup melalui penjumlahan & perkalian skalar. Sebuah submodul $M$ ditasbihkan sebagai *finitely generated* jika ia ditopang/dibangun penuh oleh sekumpulan kecil elemen skalar yang berhingga $\{m_1, \dots, m_k\}$. Ini dapat dianalogikan sebagai asumsi matematis bahwa ruang ekuilibrium regresi observasi kita "berdimensi terbatas / dapat dihitung polanya dengan akit."
2. **Ideal & *Jacobson Radical* ($J(R)$):** Pada Gelanggang $R$, **Ideal** adalah subhimpunan yang bersifat menyerap terhadap operasi perkalian dari elemen gelanggang. Irisan dari semua ideal maksimal pada gelanggang tersebut membentuk **Jacobson Radical**, $J(R)$. Dalam pemodelan regresi terapan kita, unsur di $J(R)$ merepresentasikan komponen anomali atau *noise* ekstrem; yang mana kontribusinya pada fungsi prediktif (modul *finitely generated*) dapat mendistorsi model perhitungan regresi.
3. **Modul Faktor $M/N$ (Ruang Koset):** Didefinisikan sebagai himpunan kelas ekuivalensi $x + N$. Aljabar hasil bagi ini digunakan untuk membatasi variansi dari setiap amatan regresi tunggal melalui interval toleransi konstan (*margin of limit tolerances*).

### 2.3 Himpunan dan Submodul Fuzzy
Diberikan himpunan semesta $X$. Himpunan fuzzy $A$ dalam $X$ dikarakterisasi oleh **Fungsi Keanggotaan** $\mu_A : X \to [0, 1]$. Berangkat dari hal tersebut, apabila $M$ adalah suatu $R$-modul, sebuah himpunan fuzzy $\mu : M \to [0, 1]$ disebut **Submodul Fuzzy** dari $M$ jika fungsi keanggotaannya selalu menaati properti peluruhan selisih & perkalian linear berstruktur:
1. $\mu(x - y) \ge \min\{\mu(x), \mu(y)\}$
2. $\mu(rx) \ge \mu(x)$ 

*(Ross, 2010; Al-Aeashi & Abbas, 2014)*

### 2.4 Kajian Ekstensif Jurnal: Perluasan Fuzzy Pada Aljabar Nakayama Memodelkan Anomali
Berlandaskan pada pemodelan matematika oleh Zamani (2026), pemodelan fungsional anomali pada regresi ini direpresentasikan dengan mengadopsi teorema *Fuzzy Nakayama's Lemma* dan integrasi modul abstrak komutatif.

#### A. Konstruksi dan Motivasi Lemma Nakayama
Dalam aljabar komutatif klasik, **Lemma Nakayama** menyatakan suatu properti anihilasi (*annihilation*): untuk sebuah $R$-modul $M$ bertipe *finitely generated*, bila ideal $J(R)$ memenuhi relasi dominan $J(R)M = M$, maka sistem modul tersebut bersesuaian dengan elemen trivial atau $M = 0$. Dalam kajian *Fuzzy*, batasan ini direlaksasikan: kestabilan model pengamatan dapat direpresentasikan berdasarkan *degree of truth* secara fungsional ($\mu$), yang relevan untuk menampung anomali dalam himpunan *time-series*.

#### B. Dekonstruksi Objek Nomenklatur Fuzzy
Agar properti anihilasi ini sejalan dengan pengujian data BTC/USD nantinya, jurnal merumuskan beberapa *nomenklatur matematis lanjutan*:
1. **Fuzzy Ideal ($\xi$)**: Perluasan syarat ideal ke rentang $[0, 1]$. Berfungsi membatasi operasi selisih yang harus memenuhi $\xi(a-b) \ge \min\{\xi(a), \xi(b)\}$ dan mencerminkan sifat penyerapan elemen ideal terhadap operasi perkalian komutatif $\xi(ab) \ge \max\{\xi(a), \xi(b)\}$. Model di perumusan ini merepresentasikan besaran intervensi suatu observasi.
2. **Submodul Trivial Puncak ($1_\theta$)**: Anomali fungsi trivial (analogi himpunan kosong dalam aljabar modul). Objek trivial ini merepresentasikan kondisi di mana suatu titik data tereduksi menjadi probabilitas nol seutuhnya di dunia nyata; fungsi tersebut bernilai $1_\theta(0) = 1$, sementara untuk elemen lainnya selalu bernilai $0$. Nilai ini dicapai setiap kurva data menembus batas deviasi ekstrem di $1.5\sigma$.
3. **Fuzzy Jacobson Radical ($FJ(R)$)**: $FJ(R) = \bigcap \{\tau \mid \tau \text{ ideal fuzzy non-konstan}\}$. Formulasi ini melambangkan titik irisan dari elemen-elemen ekstrem ideal. Pada analisis riil di lapangan, elemen inilah representasi dari *"Outlier Radical/Noise Pasar Ekstem"* ($> \pm 1.5\sigma$). 
4. **Hasil Kali Modular Konvolutif ($\odot$)**: Merupakan wujud *Sup-Min Composition*:  $(\xi \odot \lambda)(z) = \sup_{z = \sum r_i x_i} [\min_i (\min\{\xi(r_i), \lambda(x_i)\})]$. Mekanisme gabungan hasil kali ini menceritakan fungsi irisan di mana *ideal radikal* memengaruhi ruang pengamatan dari variabel dasar/normal regresi kombinasi linier.

#### C. Aksiomatisasi Teorema Reduksi Regresi
> **Teorema Relasi Penyerapan (Teorema Homologi).** Misalkan $M$ adalah suatu $R$-modul, $\lambda$ adalah perluasan toleransi regresi (fungsi submodul prediktif *finitely generated*), dan ada himpunan deviasi berbobot $\xi$. Jika kondisi $\xi \odot \lambda = \lambda$ terpenuhi (anomali mendominasi tren secara utuh), maka terdapat elemen $a \in R$ dengan nilai $\xi(a) = 1$ yang menstimulasi persamaan reduksi trivial: $$ 1_{N(1+a)} \odot \lambda = 1_\theta $$.
*(Anomali ekstrem mengeliminasi pola ekuilibrium tren).*

> **Fuzzy Nakayama's Lemma (Reduksi Observasi Akibat Anomali).** Misalkan di dalam data observasi riil termuat tingkat deviasi (anomali fuzzy) berskala signifikan di atas toleransi ekstrem yang memenuhi $\xi \subseteq FJ(R)$. Jika anomali ini mendominasi fungsi ruang submodul regresi secara keseluruhan ($\xi \odot \mu = \mu$), disimpulkan secara matematis bahwa tren normal tersebut tereduksi secara mutlak (anihilasi) menjadi Submodul Trivial: $$ \mu = 1_\theta $$.
*(Observasi yang melalui ekstrem radikal akan direduksi probabilitasnya menjadi nilai Trivial alias $\mu = 0$).*

## BAB III: METODOLOGI PENELITIAN

### 3.1 Konstruksi Ruang Modul dan Gelanggang
Untuk menjembatani rumusan teori di atas ke ranah terapan nyata (studi kasus anomali BTC/USD), sistem dikonstruksi secara formal:
1. **Definisi Gelanggang ($R$) dan Modul ($M$):** Ditetapkan $R = \mathbb{R}$ (Lapangan Bilangan Real yang memenuhi sifat gelanggang komutatif) sebagai himpunan skalar koefisien. Kumpulan kurva berdimensi waktu diposisikan sebagai Modul Polinomial, $M = \mathbb{R}[x]$. Formasi ini sah sebagai sebuah **$R$-Modul**.
2. **Submodul Pembangun Tren ($N \subset M$):** Ruang regresi ekuilibrium direpresentasikan sebagai *Finitely Generated Submodule* berorde tertentu (akan dicari parametrisasinya yang paling optimal berdasarkan MSE terendah). Ekspansi submodul ruang ini didefinisikan sebagai $N = \text{span}_{\mathbb{R}}\{1, x, \dots, x^n\} \subseteq M$, mendapuk polinomial prediktif $p(x) \in N$ sebagai generator kurva.
3. **Sabuk Evaluator (Fuzzy Region $\mu$):** Meluaskan himpunan tegas regresi ke dalam balutan *Fuzzy* untuk mempertahankan identitas *outlier* di ruang probabilistik.

### 3.2 Fungsi Keanggotaan pada Modul Faktor
Secara empiris, fungsi keanggotaan dikomputasikan dengan reduksi norm residu terhadap margin deviasi absolutnya ($d$):
$$ \mu(y_i) = \max\left(0, 1 - \frac{|y_i - p(x_i)|}{d} \right) $$

**Logika Aksiomatisasi Submodul berbatas Norma:**
Jika fungsi evaluasi regresi linier dioperasikan di alam bebas sebagai *Fuzzy Module* di ruang vektor bebas, ia rentan terdisrupsi aksioma skalar (*multiplicative scalar axiom* $\mu(ry) \ge \mu(y)$). Guna memastikan kepatuhan aljabarnya, himpunan terapan ini dilandaskan pada konsep ruang metrik **Modul Hasil-Bagi Berbatas Norm (Bounded Quotient Norm Module)** $M/N$.  Semesta evaluasi ini mereduksi pengamatan ke dalam fungsi ekuivalensi spasial relatif konstan pada ruang toleransi terbatas (*Bounded Support*).

* $p(x_i)$ : Elemen koset sentral (kondisi ideal regresi di mana $\mu = 1$).
* $d$ : Radius *Fuzzy Spread* ($\sigma$). Residu devian yang meluap melewati ekuator penyebaran ini niscaya dilemparkan secara logis kepada keanggotaan nol mutlak / $1_\theta$ ($\mu = 0$).
* $\mu(y_i)$ : Matriks derajat proksimitas kebenaran submodul tren fungsional.

**Justifikasi Pemilihan Fungsi Keanggotaan Segitiga Simetris:**
Pemilihan perumusan *Symmetric Triangular Membership Function* (TMF) ini, dibandingkan fungsi kurva lonceng (Gaussian) atau Sigmoid, didasarkan pada dua landasan utama:
1. **Sifat *Compact Support* (Eksistensi Nol Mutlak):** Fungsi Gaussian menurun keruncingannya secara asimtotik namun tidak pernah menyentuh nilai eksak $0$. Sebaliknya, fungsi TMF yang dibatasi oleh operator $\max(0, \dots)$ menjamin adanya margin di mana $\mu = 0$ secara absolut. Ini merupakan **pra-syarat wajib** agar fenomena operasi elemen radikal dalam mereduksi submodul seutuhnya menjadi **submodul trivial ($1_\theta$)** dapat diwujudkan secara matematis dalam pembuktian *Fuzzy Nakayama's Lemma* pada bab pengujian.
2. **Prinsip *Minimum Fuzziness*:** Dalam komputasi regresi berbasis jarak (*distance-based*), penurunan derajat secara linear (*L1-Norm*) yang direpresentasikan oleh TMF terbukti paling efisien dalam menekan total ketidakpastian anomali fluktuasi tajam dibandingkan kurva nonlinear (Chukhrova & Johannssen, 2019; Hesamian & Akbari, 2017).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import warnings
warnings.filterwarnings('ignore')

ticker_symbol = "BTC-USD"
btc_data = yf.download(ticker_symbol, period="5y", interval="1d")

df_raw = btc_data.copy().reset_index()

target_col = [c for c in df_raw.columns if 'Close' in c][0]

df_raw['Indeks_Waktu (x)'] = np.arange(1, len(df_raw) + 1)
df_raw['Harga_Tutup (y)'] = df_raw[target_col].values

jumlah_sampel = df_raw.shape[0]
print(f"Total Sampel Observasional: {jumlah_sampel}")

# Menampilkan 20 data terbaru (dari ujung akhir periode/hari ini)
display(df_raw[['Date', 'Indeks_Waktu (x)', 'Harga_Tutup (y)']].tail(20))
```

### 4.1 Koleksi Deskriptif Data Observasi

### 4.2 Evaluasi Kelayakan Derajat Polinomial Pembangun (Berdasarkan MSE)
Berdasarkan analisis pemodelan komparatif pada jurnal *Prediction of Cryptocurrency Prices Using LSTM, SVM and Polynomial Regression* (Al Giffary & Sulianta, 2024), derajat regresi polinomial mempengaruhi secara telak nilai galat aproksimasi fungsi tren. Pada hasil pengujian literatur tersebut, disoroti bahwa **derajat 6 ($\text{degree }6$)** menghasilkan nilai hampiran rata-rata galat terendah. 

Untuk menghindari *overfitting* sekaligus memaksimalkan pelacakan tren fluktuasi kripto kita pada data BTC/USD terkini, kita akan memvisualisasikan langsung dan menghitung parameter MSE (*Mean Square Error*) secara empiris atas himpunan fungsi polinomial berderajat 3 hingga 8. Derajat pemegang MSE paling optimum dari program di bawah ini kemudian akan didaulat langsung sebagai "Kriteria Ekuilibrium $p(x)$" dari sistem Fuzzy Nakayama's Lemma kita.

```python
# 1. Mengeksplorasi dan Menghitung Derajat Polinomial (Orde 3 hingga 8) dengan MSE Terendah
x_data = df_raw['Indeks_Waktu (x)'].values
y_data = df_raw['Harga_Tutup (y)'].values

derajat_uji = range(3, 9)
mse_terendah = float('inf')
derajat_terpilih = 3

x_smooth_eval = np.linspace(min(x_data), max(x_data), 1000)

fig, ax = plt.subplots(figsize=(14, 8))
ax.scatter(x_data, y_data, color='lightgray', alpha=0.4, s=6, label='Nilai Aktual BTC/USD $(y_i)$')

colors = ['crimson', 'forestgreen', 'mediumblue', 'darkviolet', 'darkorange', 'saddlebrown']

print("--- EVALUASI MEAN SQUARE ERROR (MSE) REGRESI POLINOMIAL ---")
for d, c in zip(derajat_uji, colors):
    coef = np.polyfit(x_data, y_data, d)
    p = np.poly1d(coef)
    y_pred = p(x_data)
    mse = np.mean((y_data - y_pred)**2)
    
    # Visualisasi
    ax.plot(x_smooth_eval, p(x_smooth_eval), color=c, linewidth=2, label=f'Derajat {d} (MSE: {mse:,.0f})')
    
    status = ""
    if mse < mse_terendah:
        mse_terendah = mse
        derajat_terpilih = d
        status = "  <-- REKOR TERBAIK SEMENTARA"
        
    print(f"Polinomial Derajat {d} | MSE: {mse:,.2f}{status}")

ax.set_title('Visualisasi Perbandingan Model Generator Polinomial $\\mathbb{R}[x]$ (Derajat 3 hingga 8)', fontsize=14, pad=12)
ax.set_xlabel('Hari Sekuensial Kejadian ($x$)')
ax.set_ylabel('Harga Aktual (USD, $y$)')
ax.legend(loc='upper left', frameon=True)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

print(f"\n=> KEPUTUSAN FINAL: Derajat {derajat_terpilih} dipilih sebagai struktur generator Modul p(x) otomatis (berdasarkan MSE Terendah).")
```

### 4.3 Komputasi Submodul $\mathbb{R}[x]$ dan Grafik Fungsi Toleransi
Dengan disahkannya **Derajat optimal (MSE terkecil)** di atas sebagai derajat polinomial pembangun utama $p(x)$, kita melebarkan margin penyerapan elemen sebagai visualisasi himpunan semesta (*Fuzzy Region*).

**Formulasi Lebar Submodul ($d$):** Luasan pita ekuivalensi dilarang dimanipulasi dengan tebakan spekulatif. Jarak rentang deviasi toleransi dimatangkan memakai asas probabilitas matematis (Hesamian & Akbari, 2017) lewat nilai peredaman Standar Deviasi Residu ($\sigma$). Radius ditetapkan pada koefisien $1.5\sigma$, di mana titik penyimpangan ekuivalen di luarnya dinyatakan divonis gugur paksa ke submodul anomali trivial ($1_\theta$).

```python
x_data = df_raw['Indeks_Waktu (x)'].values
y_data = df_raw['Harga_Tutup (y)'].values

derajat_polinomial = derajat_terpilih  # Menggunakan derajat terbaik dari uji MSE di atas
koef_polinomial = np.polyfit(x_data, y_data, derajat_polinomial)
polinomial_ideal = np.poly1d(koef_polinomial) 

x_smooth = np.linspace(min(x_data), max(x_data), 1000)
y_ideal_smooth = polinomial_ideal(x_smooth)

# Menghitung Margin Fuzzy secara akademis berdasar Model Tanaka (1982)
y_prediksi_aktual = polinomial_ideal(x_data)
residu_aktual = y_data - y_prediksi_aktual
standar_deviasi_residu = np.std(residu_aktual)

# Penetapan Variansi Toleransi 1.5 Sigma untuk menyerap distribusi elemen wajar
d_margin = 1.5 * standar_deviasi_residu

print(f"Standar Deviasi Residu (Sigma) : {standar_deviasi_residu:.2f} USD")
print(f"Batas Terluar Toleransi (d)    : {d_margin:.2f} USD")

fig, ax = plt.subplots(figsize=(12, 7))

ax.scatter(x_data, y_data, color='royalblue', alpha=0.35, s=6,
           label='Nilai Aktual BTC/USD $(y_i)$')

ax.plot(x_smooth, y_ideal_smooth, color='crimson', linewidth=3, 
        label=f'Fungsi Polinomial Generator $\\mathbb{{R}}[x]$ Derajat-{derajat_polinomial}')

ax.fill_between(x_smooth, y_ideal_smooth - d_margin, y_ideal_smooth + d_margin, 
                color='gold', alpha=0.3, 
                label=f'Batas Toleransi Submodul Fuzzy / Spread ($d={d_margin:.0f}$)')

ax.set_title('Evaluasi Himpunan Fuzzy Submodul atas Gelanggang Daerah Integral Polinomial $\\mathbb{R}[x]$\nDataset Volatilitas Harga Kripto Harian (5 Tahun Terakhir)', fontsize=14, pad=12)
ax.set_xlabel('Hari Sekuensial Kejadian ($x$)')
ax.set_ylabel('Rata-Rata Harga (USD, $y$)')
ax.legend(loc='lower left', frameon=True)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

### 4.4 Evaluasi Karakteristik Keanggotaan Empiris $\mu(y) \in [0, 1]$

Kinerja struktur teori Himpunan Submodul berbatas Norma dapat dikuantifikasi melalui derajat serapan dari beberapa titik sampel perpanjangan waktu ($x$).
Setiap elemen yang menyentuh presisi eksak fungsi kelampauan kurva pembangun generator $p(x)$ meniscayakan fungsi bernilai keanggotaan derajat utuh $\mu = 1$. Semakin bias dispersi deviasi titik $y_i$ hingga menembus garis batas asimptot radius toleransi, struktur afeksi *fuzzy*-nya rontok berakselerasi menuju $\mu = 0$. 

Tiga uji pemetaan parsial di blok algoritma berikut mensimulasikan hukum aksiomatik fungsional *Fuzzy Modul* ini berbasis kalkulus *L1-Norm Absolute* tanpa mencampuri library bawaan *black-box* yang kerap mengaburkan pembuktian homologi murninya.

```python
def get_fuzzy_membership(y_aktual, y_prediksi, margin_d):
    jarak_residual = abs(y_aktual - y_prediksi) 
    derajat = max(0.0, 1.0 - (jarak_residual / margin_d))
    return derajat

max_id = len(x_data) - 1
indeks_evaluasi = [300, max_id // 2, max_id - 10]

for idx in indeks_evaluasi:
    x_val = x_data[idx]
    y_aktual = y_data[idx]
    y_teori = polinomial_ideal(x_val) 
    
    mu_val = get_fuzzy_membership(y_aktual, y_teori, d_margin)
    
    print(f"\nObservasi Hari ke-{x_val:.0f} (Indeks {idx}):")
    print(f"» y_aktual : {y_aktual:.1f}")
    print(f"» y_teori  : {y_teori:.1f}")
    print(f"» residual : {abs(y_aktual - y_teori):.1f}")
    print(f"» mu_val   : {mu_val:.4f}")
```

### 4.5 Simulasi Prediksi Kripto (Forecasting) Menggunakan Generator Submodul
Karena struktur tren (*Finitely Generated Submodule*) sudah didefinisikan secara tegas melalui rumus parameter polinomial $p(x)$ beserta luasan semesta observasinya ($d\_margin$), kita dapat melakukan **ekstrapolasi/prediksi probabilitas tren** untuk beberapa hari ke depan (hari ke-$t_{last} + n$). 

Pendekatan prediksi (forecasting) menggunakan regresi polinomial ini beralasan dan selaras dengan metode penelitian komparatif yang dilakukan oleh **Al Giffary & Sulianta (2024)**. Pada kajian tersebut, regresi polinomial diterapkan sebagai instrumen prediktif untuk menilik arah pergerakan aset kripto berbasis historis, meskipun secara jangka panjang umumnya dikomparasi dengan model *deep learning* (LSTM).

Namun, perlu dibarengi sebuah *caveat* (peringatan) secara matematis: Regresi polinomial orde tinggi, meskipun kelewat akurat menekan MSE di dalam rentang sampel masa lalu (*interpolasi*), sangat rawan terhadap **fenomena Runge (*Runge's Phenomenon*) / efek osilasi ujung**. Ekstrapolasi regresi tingkat tinggi untuk jangka panjang lazimnya akan melonjak menyimpang drastis ke arah ketakhinggaan. Karenanya, agar prediksi tetap merepresentasikan *Fuzzy Region* yang realistis dan aplikatif, pemetaan perkiraan lintasan ini wajib dibatasi pada jendela waktu yang sangat pendek (contohnya **3 hari ke depan**).

Untuk memperjelas *bagaimana* prediksi ini bekerja, kita akan mem-fokuskan (*zoom-in* ekstrem) visualisasi hanya pada beberapa hari terakhir. Visualisasi ini menunjukkan bagaimana garis polinomial ditarik keluar meninggalkan data riil masa historis beserta sabuk peredam Fuzzy-nya.

```python
# Penentuan Forecasting Kripto 3 Hari ke Depan (Jangka Pendek)
hari_prediksi = 3
hari_terakhir = x_data[-1]

print(f"--- PREDIKSI HARGA SENTRAL (PUNCAK MU=1) UNTUK {hari_prediksi} HARI MENDATANG ---")
print(f"Menggunakan Fungsi Ekstrapolasi Polinomial Derajat-{derajat_terpilih}")

x_future = np.arange(hari_terakhir + 1, hari_terakhir + 1 + hari_prediksi)
y_future_ideal = polinomial_ideal(x_future)

for h, y in zip(x_future, y_future_ideal):
    # Menyajikan prediksi beserta ruang Fuzzy toleransinya
    batas_bawah = y - d_margin
    batas_atas  = y + d_margin
    print(f"Hari Ke-{h:04d} | Prediksi Harga Sentral: {y:10,.2f} USD  || Margin Toleransi [ {batas_bawah:,.2f} s/d {batas_atas:,.2f} ]")

# Memvisualisasikan bagian pinggir dari kurva dengan ZOOM EKSTREM
observasi_pinggir = 15  # Jendela observasi historis sempit agar garis proyeksi terlihat jelas melanjutnya
x_tinjauan = np.concatenate((x_data[-observasi_pinggir:], x_future))
y_tinjauan = polinomial_ideal(x_tinjauan)

fig, ax = plt.subplots(figsize=(11, 7))

# 1. Plot Fungsi Ekuilibrium (Jejak Historis)
ax.plot(x_tinjauan[:-hari_prediksi], y_tinjauan[:-hari_prediksi], color='crimson', lw=2, linestyle='-', label=f'Garis Regresi Historis (Derajat {derajat_terpilih})')

# 2. Plot Fungsi Prediksi (Ekstrapolasi Masa Depan)
# Menyambungkan titik historis terakhir dengan titik pertama masa prediksi
ax.plot(x_tinjauan[-hari_prediksi-1:], y_tinjauan[-hari_prediksi-1:], color='limegreen', lw=3, linestyle='--', marker='o', markersize=6, label='Jalur Prediksi Ekstrapolasi')

# 3. Scatter Data Observasi Riil
ax.scatter(x_data[-observasi_pinggir:], y_data[-observasi_pinggir:], color='black', alpha=0.7, s=40, label='Observasi Pasar Aktual', zorder=5)

# 4. Sorot titik pisah (Batas Saat Ini)
ax.axvline(x=hari_terakhir, color='gray', linestyle=':', label='Batas Waktu Observasi Terakhir')

# 5. Anotasi Interaktif (Teks penjelasan di dalam grafik)
y_aktual_terakhir = y_data[-1]
ax.annotate('Data Observasi Terakhir', 
            xy=(hari_terakhir, y_aktual_terakhir), 
            xytext=(hari_terakhir - 3, y_aktual_terakhir - (d_margin*0.5)),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6),
            fontstyle='italic')

ax.annotate('Arah Prediksi Polinomial\nMelebihi observasi nyata', 
            xy=(x_future[-1], y_future_ideal[-1]), 
            xytext=(x_future[-1] - 4, y_future_ideal[-1] + (d_margin*0.6)),
            arrowprops=dict(facecolor='green', shrink=0.05, width=1, headwidth=6), 
            color='darkgreen', weight='bold')

# 6. Toleransi Fuzzy (Spread Margin)
ax.fill_between(x_tinjauan, y_tinjauan - d_margin, y_tinjauan + d_margin, color='gold', alpha=0.2, label=f'Fuzzy Submodule Margin (+/- {d_margin:,.0f} USD)')

ax.set_title(f'Penjelasan Visual Metode Forecasting (Zoom Ekstrem {observasi_pinggir} Hari Terakhir + {hari_prediksi} Hari Prediksi)', fontsize=13, pad=15)
ax.set_xlabel('Indeks Waktu Observasi Harian (x)')
ax.set_ylabel('Harga BTC/USD (y)')
ax.grid(True, ls='--', alpha=0.5)
ax.legend(loc='lower left')

plt.tight_layout()
plt.show()
```

## BAB V: KESIMPULAN

Berdasarkan sintesis antara kajian aljabar abstrak dan implementasi simulasi dinamika pasar mata uang kripto (BTC/USD), dapat dikemukakan beberapa simpulan yang merefleksikan *output* komputasi proyek ini:

1. **Efektivitas Gelanggang Polinomial Berbasis Optimasi Auto (MSE):** Pemanfaatan aljabar $\mathbb{R}[x]$ dapat diterapkan secara empirik untuk mengekstraksi ekuilibrium tren dari volatilitas acak pasar kripto. Tidak ada satu orde yang selamanya benar; evaluasi komparatif (*Mean Square Error*) antar derajat 3-8 terbukti fundamental untuk menentukan Modul Faktor pembangun tren $p(x)$ (di mana Polinomial Berderajat 6 seringkali memberikan keseimbangan MSE terbaik, konsisten dengan studi regresi oleh Al Giffary & Sulianta, 2024).
2. **Fleksibilitas Submodul Fuzzy:** Berbeda dengan himpunan kaku (*crisp set*) yang menolak pengamatan begitu luput dari kurva ideal regresi, pembentukan dimensi luasan via norm berbatas memberikan rentang toleransi (*bounded support area*) menggunakan patokan standar deviasi dari volatilitas residu ($1.5\sigma$). Parameter pelebaran sabuk Fuzzy $d$ ini memfasilitasi absorpsi anomali kecil agar tetap terhitung sebagai derajat kepuasan parsial.
3. **Interpretasi Dinamis Derajat Keanggotaan ($\mu$):**
   * $\mu_i = 1$: Data berpadu tepat secara mutlak melintas pada garis generator $\mathbb{R}[x]$.
   * $0 < \mu_i < 1$: Data mendapati deviasi wajar, namun masih diikat secara probabilistik oleh ruang toleransi submodul. Terbukti dari hasil keluaran *(output)* uji evaluasi, data yang cukup lebar deviasinya dari tren tetap direngkuh, mencegah penolakan sistem seketika atas deviasi harga akibat rumor kecil pasar. 
   * $\mu_i = 0$: Data terdelineasi mutlak sebagai ekstrem tak tertolerir, tercabut menjadi anggota murni dari himpunan identitas trivial.
4. **Validasi Empiris Lemma Nakayama Fuzzy pada Kondisi Anomali (Anihilasi):** Tereduksinya fungsi keanggotaan ke tepi limitnya (nol) memvisualisasikan Lemma Nakayama Fuzzy secara kasat mata. Sesuai propertinya (*Fuzzy Jacobson Radical* yang memonopoli gelanggang dan mereduksinya menjadi submodul Trivial $1_\theta$), kejadian saat BTC menembus sabuk kuning (di atas batas asimptot $>1.5\sigma$) membuktikan bahwa *noise* intervensi masif secara otomatis meruntuhkan tingkat kecocokan historik menjadi **nol mutlak**, membersihkan fungsi prediktif dari gangguan data anomali fiktif secara presisi.
5. **Aplikasi Ekstrapolasi (*Forecasting*) & Limitasi Runge:** Fungsi *Finitely Generated Submodule* ini pada hakikatnya memuat properti prediktif (forecasting) dalam rentang waktu temporer yang sangat singkat (mis. ekstrapolasi 7 hari). Validitas metode tersebut untuk instrumen aset digital telah didukung literatur. Walau demikian, kelemahan mendasar himpunan $\mathbb{R}[x]$ berderajat tinggi ialah kerentanan atas *Runge's Phenomenon* (Osilasi Ujung), di mana ekstensi waktu untuk prediksi jangka panjang (berbulan-bulan) niscaya akan merusak ekuilibrium submodul memandu lintasan kurva ke arah distorsi harga ekstrem tak terhingga.
