```python
x = Real('x')
y = Real('y')
z = Real('z')

s = Solver()
s.add(x + y - 3*z == 1)
s.add(-2*x + y + z == 3)
s.add(3*x - y + 2*z == 4)

print(s.check())
mod = s.model()
print(mod)
```

```python
A = Int('A')
B = Int('B')
C = Int('C')
D = Int('D')
E = Int('E')

var = [A, B, C, D, E]

s = Solver()
s.add(((1000*A + 100*B + 10*C + D)*E) == (1000*D + 100*C + 10*B + A))
s.add(And(A >= 1, B >= 0, C >= 0, D >= 1, E >= 1))
s.add(And(A <= 9, B <= 9, C <= 9, D <= 9, E <= 9))
s.add(Distinct(A, B, C, D, E))

print(s.check())
mod = s.model()
print(mod)
```

```python
F = Int('F')
vars = [A, B, C, D, E, F]

s = Solver()
s.add(Distinct(vars))
s.add(*[v >= 1 for v in vars])
s.add(*[v <= 6 for v in vars])

s.add(Not(vars[0] == 1))
s.add(Not(vars[0] == 2))

s.add(Not(vars[1] == 1))

s.add(Not(vars[2] == 4))
s.add(Not(vars[2] == 5))

s.add(Not(vars[3] == 4))
s.add(Not(vars[3] == 5))

num = 0
while s.check() == sat:
    mod = s.model()
    num += 1
    print(f"Solution {num}:")
    print(mod)
    s.add(Or([v != mod[v] for v in vars]))
```
